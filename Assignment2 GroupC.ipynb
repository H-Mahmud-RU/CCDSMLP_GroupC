```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/H-Mahmud-RU/CCDSMLP_GroupC/main/Survey-Cleaning.csv'
df = pd.read_csv(url)
```

```python
df.isnull().values.sum()
```

```python
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
cnt=0
for row in df['Area']:
    try:
        int(row)
        df.loc[cnt, 'Area']=np.nan
    except ValueError:
        pass
    cnt+=1
df
```

```python
cnt=0
for row in df['Device']:
    try:
        int(row)
        df.loc[cnt, 'Device']=np.nan
    except ValueError:
        pass
    cnt+=1
df
```

```python
df['InternetCost'] = pd.to_numeric(df['InternetCost'], errors='coerce')
df['FamilyIncome'] = pd.to_numeric(df['FamilyIncome'], errors='coerce')
df
```

```python
df['Area'].fillna(df['Area'].mode()[0], inplace=True)
df['InternetCost'].fillna(df['InternetCost'].mode()[0], inplace=True)
df['Device'].fillna(df['Device'].mode()[0], inplace=True)
df
```

```python
df.groupby('Area')['FamilyIncome'].median()
```

```python
df['FamilyIncome'] = df['FamilyIncome'].fillna(df.groupby('Area')['FamilyIncome'].transform('median'))
df['FamilyIncome'] = df['FamilyIncome'].fillna(df['FamilyIncome'].median())
df
```

```python
df.isnull().values.any()
```

```python
df.InternetCost = df.InternetCost.astype('int64') 
df.FamilyIncome = df.FamilyIncome.astype('int64') 
df
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
df['Gender']= le.fit_transform(df['Gender']) 
df['Area']= le.fit_transform(df['Area']) 
df['Device']= le.fit_transform(df['Device'])
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df)
scaled = pd.DataFrame(scaler.transform(df),columns=df.columns)
scaled
```
